```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("adult.csv")

print("========== FIRST 5 RECORDS ==========")
print(df.head())

print("\n========== DATASET INFO ==========")
print(df.info())

print("\n========== SHAPE ==========")
print(df.shape)

print("\n========== COLUMN NAMES ==========")
print(df.columns)

print("\n========== SUMMARY ==========")
print(df.describe())

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

# Replace '?' with NaN
df.replace("?", np.nan, inplace=True)

# Fill Missing Values
df.fillna(method="ffill", inplace=True)

print("\n========== AFTER CLEANING ==========")
print(df.isnull().sum())

# Remove Duplicates
df.drop_duplicates(inplace=True)

print("\nRows after removing duplicates:", df.shape)

# -----------------------------
# Data Integration (Demo)
# -----------------------------
part1 = df.iloc[:1000]
part2 = df.iloc[1000:]
integrated_data = pd.concat([part1, part2], ignore_index=True)

print("\nIntegrated Dataset Shape:", integrated_data.shape)

# -----------------------------
# Data Transformation
# -----------------------------
df["hours-per-week"] = df["hours-per-week"] * 2

print("\nData Transformation Completed")

# -----------------------------
# Outlier Detection
# -----------------------------
plt.figure(figsize=(6,4))
plt.boxplot(df["age"])
plt.title("Box Plot of Age")
plt.show()

# -----------------------------
# Income Distribution
# -----------------------------
plt.figure(figsize=(6,4))
df["income"].value_counts().plot(kind="bar")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()

# -----------------------------
# Age Distribution
# -----------------------------
plt.figure(figsize=(6,4))
plt.hist(df["age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# Gender Distribution
# -----------------------------
plt.figure(figsize=(6,6))
df["gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()

# -----------------------------
# Education Distribution
# -----------------------------
plt.figure(figsize=(8,5))
df["education"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Education Levels")
plt.xlabel("Education")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# Workclass Distribution
# -----------------------------
plt.figure(figsize=(8,5))
df["workclass"].value_counts().head(10).plot(kind="bar")
plt.title("Top Workclass")
plt.xlabel("Workclass")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Save Dataset
df.to_csv("adult_cleaned.csv", index=False)

print("\nDataset Saved Successfully as adult_cleaned.csv")
print("\nPROJECT COMPLETED SUCCESSFULLY")
```
